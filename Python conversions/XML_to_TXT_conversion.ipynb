```python
import warnings
warnings.filterwarnings("ignore")

# Import importTrackmateTracks_function
%run importTrackmateTracks_function.ipynb

## File paths
filename = r'C:\Users\ryanr\OneDrive\Desktop\Krapf Research\RNA data analysis\Trajectories\Test 2024-3-30\Dish 1\Sample1001_control_Tracks.xml'
savefile = r'C:\Users\ryanr\OneDrive\Desktop\Krapf Research\RNA data analysis\Trajectories\Test 2024-3-30\Dish 1\Sample1001_control_Tracks.txt'

# Parameters
pixelsize = 0.130  # Pixel size in micrometers
Max = 1500  # Maximum track length in frames

# Import tracks
tracks, metadata = import_trackmate_tracks(filename)

# Initialize empty matrix
    # len(tracks) = number of trajectories
    # len(tracks[0]) = length of first trajectory
M, N = len(tracks), len(tracks[0])
B = [[float('NaN') for _ in range(2 * M)] for _ in range(Max)]
format_str = ''
title_str = ''
traj_counter = 0

# Populate matrix and format strings
for i, track in enumerate(tracks):
    for j, point in enumerate(track):
        B[j][2 * i] = point[0] * pixelsize  # x coordinate
        B[j][2 * i + 1] = point[1] * pixelsize  # y coordinate
    format_str += '%12.4f\t%12.4f\t'
    title_str += f'      X{traj_counter}\t      Y{traj_counter}\t'
    traj_counter += 1

# Write to file
with open(savefile, 'w') as f:
    f.write(title_str + '\n')
    for row in B:
        f.write(format_str % tuple(row) + '\n')
    f.close()
```
